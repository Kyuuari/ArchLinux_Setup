# Import Data
White Wine Data taken from https://archive.ics.uci.edu/ml/datasets/Wine+Quality 

```python
import pandas as pd

#import data
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv'
dataset = pd.read_csv(url, sep='[;]', engine='python')
dataset.columns = [col[1:-1] for col in dataset.columns]

dataset.head(5)
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
dataset = dataset.astype({"quality":'category'})
dataset.info()
```

```python
data = dataset.sample(frac=0.989, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
data['quality'].value_counts()
```

# Enviroment Setup
Setup PyCaret Enviroment

```python
from pycaret.classification import *
wine_env = setup(data = data, target = 'quality', session_id=123) 
```

# Find Best Models

use the comepare_models() function to determine the best models

```python
best = compare_models(n_select=3)
```

## Create Top 3 Models
Create the models after selecting the top 3 performers,
top 3 performers are based on accuracy

```python
et = create_model('et')
```

```python
print(et)
```

```python
rf = create_model('rf')
```

```python
lgm = create_model('lightgbm')
```

## Tune Models

```python
tuned_et = tune_model(et)
```

```python
tuned_rf = tune_model(rf)
```

```python
tuned_lgm = tune_model(lgm)
```

# Evalutate Models

```python
evaluate_model(tuned_et)
```

```python
evaluate_model(tuned_rf)
```

```python
evaluate_model(tuned_lgm)
```

## Results

After looking through the confusion matrix of all the models we can see that both the et and rf perform quite simularily however the extra trees classifier performs a little bit better in terms of accuracy. One thing to also note is that when looking at the feature importance for both the et and rf show that the biggest factors to determine quality is the alcohol, volatile acidity, and free sulfur dioxide amounts.
